```
sns.set(style='darkgrid')
fig=plt.gcf()
fig.set_size_inches(14,10.5)
values = pd.DataFrame(dataset['State'].value_counts()[:10])
count=0
plt.bar(values.index,values['State'],color=['red','orange','blue','white','grey','grey','grey','grey','grey','grey'])
plt.xticks(rotation=90,fontsize=10)
plt.title('No, of cases in states',fontsize=20)
plt.xlabel('States',fontsize=15)
plt.ylabel('Cases',fontsize=14)
plt.show()
```
